```python
import nltk
```

```python
# Opening File
file_content = open("input.txt").read()
```

```python
file_content
```

```python
# Splitting phrases
phrases = list(file_content.split('\n'))
#print('\n'.join(phrases))

v = phrases[:]
```

```python
v
```

```python
# Changing case to lowerCase
tmp = []

for phrase in phrases:
    tmp.append(phrase.lower())

phrases = tmp
```

```python
phrases
```

```python
# Removing Symbols
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')

tmp = []

for phrase in phrases:
    tmp.append(tokenizer.tokenize(phrase))
    
phrases = tmp
```

```python
phrasesTemp = phrases[:]
```

```python
d2 = {}

def check(ip1,ip2):
    count = 0
    for i in ip1:
#         print(i,' ',ip2)
#         print([tp[:len(i)] for tp in ip2])
#         print([tp[-len(i):] for tp in ip2])
        if i in ip2:
            count+=1
        elif i in [tp[:len(i)] for tp in ip2]:
#        elif i in ip2[:][:len(i)]:
            count+=1
        elif i in [tp[-len(i):] for tp in ip2]:
#        elif i in ip2[:][:len(i):-1]:
            count+=1
    print(count)
    if count==len(ip1):
        return True
    else:
        return False

sentences = phrases[:]    
sentences3 = phrases[:]
```

```python
d2
```

```python
check(['ie'],['hiiie','byeee'])
```

```python


for phraseCh in phrases:
    for phrase in phrases:
        if phraseCh != phrase:
            print(phraseCh,'  ',phrase)
            if check(phraseCh,phrase):
                print('check passed')
                if phraseCh in sentences:
                    sentences.remove(phraseCh)
                    print(phraseCh,'  ',phrase)
                    if tuple(phraseCh) not in d2:
                        d2[tuple(phraseCh)]=phrase
                    elif len(d2[tuple(phraseCh)])<len(phrase):
                        d2[tuple(phraseCh)]=phrase
```

```python
d2
```

```python
indexing = {}

for j,i in enumerate(phrasesTemp):
    if i in d2.values():
        indexing[tuple(i)]=j
```

```python
f = open('output.txt','w')
for i,phrase in enumerate(sentences3):
    if tuple(phrase) not in d2:
        f.write(v[i]+'\n')
    else:
        f.write(v[i]+' - '+v[indexing[tuple(d2[tuple(phrase)])]]+'\n')
f.close()
```
